# Data Collection

```python
def get_finance_data(stock):
    s = yf.Ticker(stock)
    
    df = s.history(start = '2009-01-31', end='2021-10-20', interval = "1d")
    df = df.drop(['Dividends', 'Stock Splits'], axis=1)

    return df
```

```python
msft = get_finance_data('MSFT')
amzn = get_finance_data('AMZN')
tsla = get_finance_data('TSLA')
goog = get_finance_data('GOOG')
googl = get_finance_data('GOOGL')
nvda = get_finance_data('NVDA')
fb = get_finance_data('FB')
pypl = get_finance_data('PYPL')
nflx = get_finance_data('NFLX')
```

```python
tickers = ['fb', 'goog', 'msft', 'aapl', 'tsla', 'googl', 'nvda', 'amzn', 'pypl', 'nflx']
stock_data = yf.download(tickers, start="2009-01-31", end="2021-10-20")
```

```python
stock_data['Close'].iplot(title='Stock Close Prices', xTitle='Dates',yTitle='Close Prices')
```
